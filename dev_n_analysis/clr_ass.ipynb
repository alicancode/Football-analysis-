```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_path = r'C:\Users\alima\OneDrive\Desktop\Input_videos\output_videos\cropped_image.jpg'
img = cv2.imread(image_path)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
image_half = image[0:int(image.shape[0]/2),:]
plt.imshow(image_half)
plt.show()
```

```python
image_array = image_half.reshape(-1,3)

print(image_array.shape)
```

```python
#doing k-means clustering yay
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans.fit(image_array)

labels = kmeans.labels_

label_og = labels.reshape(image_half.shape[0], image_half.shape[1])
plt.imshow(label_og)

```

```python
corner_clusters = [label_og[0,0], label_og[0,-1], label_og[-1,0], label_og[-1,-1]]
non_player_clusters = max(set(corner_clusters), key = corner_clusters.count)
print(non_player_clusters)
```

```python
player_clusters = 1+non_player_clusters
kmeans.cluster_centers_[player_clusters]
```
